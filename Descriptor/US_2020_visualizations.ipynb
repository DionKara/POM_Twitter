### PCA Visualizations (163 days)

```python
features = ['bias-1', 'positive-1', 'figurative-0', 'offensive-1']
target = ['created_at']
```

#### Dems

```python
dem = pd.read_csv("Time_Series/joebiden_time_series.csv")
dem.sort_values(by=['created_at'], inplace=True)

dem.head()
```

```python
#STANDARDIZE THE DATA
x = dem.loc[:, features].values
y = dem.loc[:,target].values

x = StandardScaler().fit_transform(x)
```

```python
#PCA PROJECTION TO 2D
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
                           , columns = ['PC1', 'PC2'])
finalDf = pd.concat([principalDf, dem[target]], axis = 1)
finalDf
```

```python

x = principalDf['PC1'].values
y = principalDf['PC2'].values

u = np.diff(x)
v = np.diff(y)
pos_x = x[:-1] + u/2
pos_y = y[:-1] + v/2
norm = np.sqrt(u**2+v**2) 


plt.rc('font', size=11)
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(x,y, marker="o", markersize=8, markerfacecolor='tab:orange')
ax.plot(-1.940779, -0.742544, marker="X", markersize=16, markerfacecolor='tab:green')
ax.plot(2.535954, 0.540342, marker="X", markersize=16, markerfacecolor='tab:red')

ax.quiver(pos_x, pos_y, u/norm, v/norm, angles="xy", zorder=1, pivot="mid", width=0.0035)

per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)




ax.annotate("Start", fontsize=15, xytext=(-1.5,-0.75), xy=(-1.940779, -0.742544), color='tab:green', arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("End", fontsize=15, xytext=(1.8,0.5), xy=(2.535954, 0.540342), color='tab:red', arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

#Add the events annotations
ax.annotate("Final Debate\n(Oct 22-23)", fontsize=15, xytext=(-0.5,1), xy=(-1.652848, 1.232088), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Election day\n(Nov 3-4)", fontsize=15, xytext=(2.7,-0.22), xy=(4.253724, -0.502425), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Media call Biden as the\n2020 presidential winner (Nov 7-7)", fontsize=15, xytext=(1.5,1), xy=(4.069944, 0.797679), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))


ax.set_title('Dems_Mean PCA 163 days   -  (Event date - Twitter reaction date)')
ax.set_xlabel('PC1 - {0}%'.format(per_var[0]))
ax.set_ylabel('PC2 - {0}%'.format(per_var[1]));
```
