### PCA Visualizations (163 days)

```python
features = ['bias-1', 'positive-1', 'figurative-0', 'offensive-1']
target = ['created_at']
```

#### Dems

```python
dem = pd.read_csv("New time series/Dem_time_series_MEAN.csv")

dem.rename(columns={'Unnamed: 0': 'created_at'}, inplace=True)

dem.head()
```

```python
#STANDARDIZE THE DATA
x = dem.loc[:, features].values
y = dem.loc[:,target].values

x = StandardScaler().fit_transform(x)
```

```python
#PCA PROJECTION TO 2D
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
                           , columns = ['PC1', 'PC2'])
finalDf = pd.concat([principalDf, dem[target]], axis = 1)
finalDf
```

```python

x = principalDf['PC1'].values
y = principalDf['PC2'].values

u = np.diff(x)
v = np.diff(y)
pos_x = x[:-1] + u/2
pos_y = y[:-1] + v/2
norm = np.sqrt(u**2+v**2) 


plt.rc('font', size=11)
fig, ax = plt.subplots(figsize=(15, 10))

ax.plot(x,y, marker="o", markersize=8, markerfacecolor='tab:orange')
ax.plot(-0.623685,0.904659, marker="X", markersize=16, markerfacecolor='tab:green')
ax.plot(-1.833898,-0.700385, marker="X", markersize=16, markerfacecolor='tab:red')

ax.quiver(pos_x, pos_y, u/norm, v/norm, angles="xy", zorder=1, pivot="mid", width=0.0035)

per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1)


ax.annotate("Start", fontsize=15, xytext=(-2,2), xy=(-0.623685,0.904659), color='tab:green', arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("End", fontsize=15, xytext=(-3,-2), xy=(-1.833898,-0.700385), color='tab:red', arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Leaks for the\ncandidates (Oct 7-8)", fontsize=15, xytext=(-3.3,3), xy=(-2.97014,-0.0412254), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Election day (Nov 8-9)", fontsize=15, xytext=(4.8,0), xy=(6.172,-1.22692), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Stein proposed\n recount (Nov 23-24)", fontsize=15, xytext=(4.8,2.2), xy=(4.99796,1.08052), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Result formally\n announced (Jan 6-9)", fontsize=15, xytext=(-1,6), xy=(-2.6839,7.01306), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))

ax.annotate("Rallies against Trump (Feb 20-20)", fontsize=15, xytext=(3.7,-2.5), xy=(5.9696,-1.31711), arrowprops=dict(arrowstyle='->,head_length=0.6,head_width=0.3'))




ax.set_title('Dems_Mean PCA 163 days   -  (Event date - Twitter reaction date)')
ax.set_xlabel('PC1 - {0}%'.format(per_var[0]))
ax.set_ylabel('PC2 - {0}%'.format(per_var[1]));
```
